```python
disableds = {'type': ['Comment Disabled', 'Rating Disabled', '0 likes'], 
             'count': [comment_disabled.video_id.count(), 
                        rating_disabled.video_id.count(),
                        v_w_0_likes.video_id.count()]}

disabled_df = pd.DataFrame(disableds)
ax = disabled_df.plot(kind='bar', title ="", rot=0, figsize=(10, 7), color='y')
ax.set_xticks(disabled_df.index)
ax.set_xticklabels(disabled_df.type)

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), ha='center', va='bottom')
    
plt.xlabel('')
plt.ylabel("Count")
# plt.ylim(0,350)
plt.show()
```

## Visualize each day's view

```python
from operator import itemgetter

df_ = df.reset_index()
date_view = df_[['trending_date', 'views']].values
date_view_counts = {}
for i in date_view:
    if i[0] not in date_view_counts:
        date_view_counts[i[0]] = i[1]
    else:
        date_view_counts[i[0]] += i[1]
dv_df = pd.DataFrame.from_dict(date_view_counts, orient = 'index')
dv_df.plot(rot=0, figsize=(20, 10), title="Total Views of Each Day")
plt.xlabel("Date (Year.Day.Month)")
plt.ylabel("Views")
```
