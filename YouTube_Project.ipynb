```python
from operator import itemgetter

tag_counts = {}
for row in df_ca['tags'][:17555]:
    for tag in row.split('|'):
        if tag not in tag_counts:
            tag_counts[tag] = 1
        else:
            tag_counts[tag] += 1
            
# Sort data by key descendingly
sorted_counts = sorted(tag_counts.items(), key=itemgetter(1), reverse=True)
top_25 = sorted_counts[:4389]# top 25%
top_20 = sorted_counts[:20]
sorted_counts_df = pd.DataFrame(top_20)
sorted_counts_df.pivot(columns=1)
counts = {'tag' :sorted_counts_df[0], 'count': sorted_counts_df[1]}
counts_df = pd.DataFrame(counts)

ax = counts_df.plot(kind='bar', title ="Most Popular Tags - Canada", rot=30, figsize=(15, 10))
ax.set_xticks(counts_df.index)
ax.set_xticklabels(counts_df.tag)
plt.xlabel('Tag')
plt.ylabel("Count")
plt.show()
```

```python
# Category id count 
category_id_count = {}
for row in df_ca['category_id']:
    if row not in category_id_count:
        category_id_count[row] = 1
    else:
        category_id_count[row] += 1
sorted_id_count = sorted(category_id_count.items(), key=itemgetter(1), reverse=True)
sorted_id_count[:6] #id counts that are over 1000
```

```python
top_categories = ['Entertainment', "People & Blogs", "News & Politics", "Comedy", "Music", "Sports"]

top_categories_df = pd.DataFrame(top_categories)
categories = {'category' :top_categories, 'count': [5577, 1858, 1817, 1693, 1582, 1162]}
categories_df = pd.DataFrame(categories)

ax = categories_df.plot(kind='bar', title ="Most Popular Categories in Canada", rot=0, figsize=(15, 10))
ax.set_xticks(categories_df.index)
ax.set_xticklabels(categories_df.category)
plt.xlabel('Category')
plt.ylabel("Count")
plt.show()
```

```python
def filter_data_df(_df, attr, value=None):
    if value is None:
        result = _df[attr]
    result = _df[(_df[attr] == value)]
    return result

def filter_data_over(_df, attr, value):
    result = _df[(_df[attr] >= value)]
    return result

def filter_data_under(_df, attr, value):
    result = _df[(_df[attr] < value)]
    return result
```
